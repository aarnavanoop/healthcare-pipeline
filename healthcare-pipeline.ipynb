Our label column is the "Diagnosis of Heart Disease" column. The values we have fall under a range(0 - 4) while the data we feed the machine learning model (MEG) should be binary so this is something to consider while cleaning the data

```python
df['Diagnosis of Heart Disease'].value_counts()
```

Step 1 — Handle object columns and nulls (do these together, order matters)

Initially I tried to replace the null values with the median but this was not possible and I ran into a type error since pandas can't find the median of an object. Hence I converted the columns to type float first then found the median to replace the null values.

```python
df['Number of Major Vessels'] = pd.to_numeric(df['Number of Major Vessels'], errors = 'coerce')
df['Thal'] = pd.to_numeric(df['Thal'], errors = 'coerce')
df['Number of Major Vessels'] = df['Number of Major Vessels'].fillna(df['Number of Major Vessels'].median())
df['Thal'] = df['Thal'].fillna(df['Thal'].median())
df.info()
df.isnull().sum()
```

Step 2 involved fixing the target column where we mapped the values so that our model would not get confused with the range of values with the heart-disease diagnosis values. Instead we make them binary, mapping the '0' value in our table to a new 0 value indicated no heart disease diagnosis and anything from 1 - 4 being mapped to a value of '1', indicating heart disease diagnosis.

```python
replacement_mapping = {0: 0,
                       1: 1,
                       2: 1,
                       3: 1,
                       4: 1}

df['Diagnosis of Heart Disease'] = df['Diagnosis of Heart Disease'].replace(replacement_mapping)
df['Diagnosis of Heart Disease'].value_counts()
```

The class ratio is balanced with a value of 1.17 so we don't have to implement Synthetic Minority Oversampling Technique

Step 3 is performing one-hot encoding on the categorical columns so that we can have them as 0s and 1s. This way our model won't treat the original data as a range from 0 - 4, treating 4 as better than 0, but rather now we have just all of the categories in their own columns.

```python
columns_to_encode = ['Chest-Pain Type', 'Resting electrocardiographic results', 'Thal', 'Slope of the peak exercise ST Segment' ]
df = pd.get_dummies(df, columns = columns_to_encode, drop_first = True, dtype = int)
```

Step 4 involved removing outliers in our continuous columns which are Serum Cholestrol and Resting Blood Pressure. I used np.log1p as the data might have zeroes and np.log(0) would give me an error. The max values below reflect the result of running the logarithmic function, reducing the outliers we have.

```python
df['Serum Cholestrol in md/dl'] = np.log1p(df['Serum Cholestrol in md/dl'])
df['Resting Blood Pressure'] = np.log1p(df['Resting Blood Pressure'])
print(df[['Resting Blood Pressure', 'Serum Cholestrol in md/dl']].head())
df.describe()
```

In step 5 we normalised the columns that have continuous and discrete data, as we have already encoded the categorical data, replaced the values in our target column to follow that of a binary classification and performed log functions to remove outliers.
