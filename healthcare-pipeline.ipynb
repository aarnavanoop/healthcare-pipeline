Now we begin the process of transforming the data so that it is suitable for our ML model.
1) We have two columns with object values that need to be converted into floats, these are also our columns with null values
   a) Handle the null values (Replace dont drop since we don't have a lot of data to begin with)
   b) Convert the column into floating point values

2) Fix the target column

   a) We need a binary target column as we want to flag a patient as at-risk or not. 0 stays as 0 and 1-4 are classified as 1.
3) Handle categorical columns

   a)We have columns like chest pain, rated from 1 - 4, the model may misinterpret this as 4 being 4 times as great as 1 when in fact we have different chest pain types not specifically a rating of chest pain. This can be handled through one hot encoding

4) Normalise the numeric columns

   a) We have values in some columns ranging from 0 - 272 while in others the max value is only 4. This can heavily skew and dominate the data hence we should normalise the data and keep it in a range of 0 - 1 for all the columns.

Determining which columns are continuous and categorical, we only normalise the continuous data and one-hot encode the categorical one
